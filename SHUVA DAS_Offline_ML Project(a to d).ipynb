###b.Find the average price of laptops for each brand. Identify which brand has the highest and lowest average price.

```python
df.info()
average_price_each_brand = df.groupby('Company')['Price_euros'].mean()
print(average_price_each_brand)
min_average_price_brand = df.groupby('Company')['Price_euros'].mean().idxmin(), df.groupby('Company')['Price_euros'].mean().min()
print(min_average_price_brand)
max_average_price_brand = df.groupby('Company')['Price_euros'].mean().idxmax(), df.groupby('Company')['Price_euros'].mean().max()
print(max_average_price_brand)
```

#c.Present the correlation between Price_euros and other numeric features such as CPU_freq, Ram, Inches, and Weight. Identify which features are most correlated with price.

```python
df.head()
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print(numeric_columns)
```

```python
data = df.select_dtypes(include=[np.number])
print(data)
plt.boxplot(data)
plt.show()
```

```python
coloumns_for_correlation = ['Price_euros','CPU_freq', 'Ram', 'Inches', 'Weight','ScreenW','ScreenH', 'PrimaryStorage', 'SecondaryStorage']
correlation_matrix = df[coloumns_for_correlation].corr()
print(correlation_matrix)


plt.figure(figsize=(5, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('correlation between Price_euros and other numeric features')
plt.show()

correlation_with_price = correlation_matrix['Price_euros'].sort_values(ascending=False)
print(correlation_with_price)
##Ram is the most positively correlated with price (i.e., 0.74)
```
